```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.columns
```

```python
df.rename(columns={'year' : 'year', 'per capita income (US$)' : 'income'}, inplace=True)
```

Rename the columns name

```python
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df.year, df.income, color='purple', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2016]])
```

```python
reg.coef_
```
